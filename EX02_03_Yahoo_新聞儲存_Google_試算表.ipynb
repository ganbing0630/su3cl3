```python
import feedparser

# Yahoo 新聞 RSS 的網址
rss_url = "https://tw.news.yahoo.com/rss"

# 使用 feedparser 解析 RSS
feed = feedparser.parse(rss_url)

# 迭代處理新聞項目
for entry in feed.entries:
    title = entry.title  # 取得新聞標題
    summary = entry.summary  # 取得新聞摘要
    link = entry.link  # 取得新聞連結

    # 印出或儲存新聞資訊
    print(f"標題：{title}")
    print(f"摘要：{summary}")
    print(f"連結：{link}")
    print("-" * 20)  # 分隔線
```

```python
from google.colab import auth
from google.auth import default
import gspread

auth.authenticate_user() # 進行授權
creds, _ = default() # 使用 default 方法獲取驗證信息
gc = gspread.authorize(creds) # 使用 creds 變數授權 gspread 模組
```

```python
import feedparser

# Yahoo 新聞 RSS 的網址
rss_url = "https://tw.news.yahoo.com/rss"

# 使用 feedparser 解析 RSS
feed = feedparser.parse(rss_url)

# 開啟 Google 試算表
url = 'https://docs.google.com/spreadsheets/d/1EeYpUzupSkvwfp3LDGYikrMNmonLJZOAtduBv1gs87k/edit?gid=0#gid=0'
workbook = gc.open_by_url(url)
# 獲取工作簿中索引為 0 的工作表，並將其儲存在 sheet 變數中
sheet = workbook.get_worksheet(0)
# 新增一列資料
sheet.append_row(['title','summary','link'])
for entry in feed.entries:
  sheet.append_row([entry.title,entry.summary,entry.link])
```
